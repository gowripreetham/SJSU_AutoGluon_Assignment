```python
# ============================================================
# AutoGluon Tabular - Automatic Feature Engineering
# Part: B.i.3 of AutoGluon Assignment
# ============================================================

!pip -q install -U pip
!pip -q install -U autogluon

from autogluon.tabular import TabularDataset, TabularPredictor
from autogluon.features.generators import AutoMLPipelineFeatureGenerator
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.datasets import make_regression
```

```python
# ============================================================
# 1Ô∏è‚É£ Create a small sample dataset
# ============================================================
# We'll create a dataframe with mixed feature types
X, y = make_regression(n_samples=100, n_features=3, random_state=1)
df = pd.DataFrame(X, columns=['A', 'B', 'C'])
df['label'] = y

# Add datetime, categorical, and text columns
df['date'] = pd.date_range('2020-01-01', periods=100)
df['category'] = np.random.choice(['Low', 'Medium', 'High'], 100)
df['text'] = np.random.choice(['Great product', 'Needs improvement', 'Average item'], 100)

print("‚úÖ Example mixed dataset created:")
df.head()
```

```python
# ============================================================
# 2Ô∏è‚É£ Apply AutoGluon Feature Engineering
# ============================================================
print("\nüîç Applying AutoMLPipelineFeatureGenerator...")
auto_gen = AutoMLPipelineFeatureGenerator()
df_transformed = auto_gen.fit_transform(df)

print("\n‚úÖ Transformed dataset (after automatic feature engineering):")
print(df_transformed.head())
print("\nüìä Original columns:", list(df.columns))
print("üìä Engineered columns:", list(df_transformed.columns))
```

```python
# ============================================================
# 3Ô∏è‚É£ Train a quick model on the engineered data
# ============================================================
print("\nü§ñ Training a quick model using engineered features...")
predictor = TabularPredictor(label='label', problem_type='regression').fit(df, time_limit=60)
results = predictor.leaderboard(df, silent=True)
results.head()
```

```python
# ============================================================
# 4Ô∏è‚É£ Demonstrate Missing Value Handling
# ============================================================
print("\nüß© Testing missing value handling...")
df_missing = df.copy()
df_missing.iloc[0] = np.nan   # make one row NaN

# Create a new feature generator for missing value test
auto_gen_missing = AutoMLPipelineFeatureGenerator()
df_filled = auto_gen_missing.fit_transform(df_missing)

print(df_filled.head())
```
