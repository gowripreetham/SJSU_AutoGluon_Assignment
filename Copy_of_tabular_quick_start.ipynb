```python
!pip -q install -U pip
!pip -q install -U autogluon
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
# ============================================================
# 1Ô∏è‚É£ Load the Titanic dataset
# ============================================================
train_data = TabularDataset("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

print("‚úÖ Loaded Titanic dataset successfully!")
print("Shape of training data:", train_data.shape)
train_data.head()
```

```python
# ============================================================
# 2Ô∏è‚É£ Define the label column (target variable)
# ============================================================
label = 'Survived'   # Target column for classification
print("\nüéØ Label column:", label)
print("Unique classes in label:", train_data[label].unique())
```

```python
# ============================================================
# 3Ô∏è‚É£ Train AutoGluon model
# ============================================================
# AutoGluon automatically handles preprocessing, feature engineering,
# model selection, and ensembling for you.

predictor = TabularPredictor(
    label=label,
    problem_type='binary',
    eval_metric='accuracy',
    path='AutogluonModelsTitanic/'
).fit(train_data, time_limit=60)
```

```python
# ============================================================
# Create or reload test data (if not already defined)
# ============================================================
from sklearn.model_selection import train_test_split

# If you split earlier, re-use the same split variable name:
# But if not, define it fresh here:
import pandas as pd

# Reload the dataset
titanic_data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Split it into train/test again
train_data, test_data = train_test_split(titanic_data, test_size=0.2, random_state=42)

print("‚úÖ Test data reloaded and defined!")
print("Test data shape:", test_data.shape)
test_data.head()
```

```python
# ============================================================
# 4Ô∏è‚É£ Evaluate model on test data
# ============================================================
print("\nüîç Evaluating on test data...")
results = predictor.evaluate(test_data, silent=True)
print("\nüìä Evaluation results:", results)
```

```python
# ============================================================
# 5Ô∏è‚É£ Make predictions (optional)
# ============================================================
print("\nüîÆ Example predictions:")
preds = predictor.predict(test_data)
preds.head()
```

```python
# ============================================================
# 6Ô∏è‚É£ Show leaderboard of trained models
# ============================================================
print("\nüèÜ Model Leaderboard:")
leaderboard = predictor.leaderboard(test_data, silent=True)
leaderboard
```
